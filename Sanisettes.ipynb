```python
import pandas as pd
import requests, json
import urllib.parse
import numpy as np

df = pd.read_csv("sanisettesparis.csv", delimiter=";")
```

```python

df["HORAIRE"].replace({"6 h - 1 h": "6h/1h", "24 h / 24": "24h/24h", "6 h - 22 h":"6h/22h", "24h/24":"24h/24h","19 h - 7 h":"19h/7h", "10h Ã  22h":"10h/22h"}, inplace=True)
```

```python
df.RELAIS_BEBE.fillna("Non", inplace=True)
df.ACCES_PMR.fillna("Non", inplace=True)
```

```python
list_arr_nan = list(df.ARRONDISSEMENT[df.ARRONDISSEMENT.isnull()].index)

list_arr_nan
```

```python
df.ARRONDISSEMENT = df.ARRONDISSEMENT.astype(str)
```

```python
for i in list_arr_nan:
    url = "https://api-adresse.data.gouv.fr/reverse/?lon=" + str(df.geo_point_2d[i].split(",")[1]) + "&lat=" + str(df.geo_point_2d[i].split(",")[0])
    print(url)
    r = requests.get(url)
    postcode = r.json()["features"][0]["properties"]["postcode"]
    print(postcode)
#     df.loc[[i]].ARRONDISSEMENT = postcode
#     df.loc[[i]]['ARRONDISSEMENT'] = df.loc[[i]]['ARRONDISSEMENT'].replace(['nan'],postcode)
    df.at[i, "ARRONDISSEMENT"]= postcode
```

```python
df.ARRONDISSEMENT = df.ARRONDISSEMENT.astype(float).astype(int)
```
